**Preprocessing on Forest Fire dataset**

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# data = pd.read_csv('forestfires.csv')
data = pd.read_csv('/content/drive/MyDrive/dataset/forestfires.csv')
```

```python
#dropping null values
data = data.dropna()
```

```python
print(data.shape)
data.head()
```

```python
data.describe()
```

```python
#Find the duplicates
data.duplicated().sum()
```

```python
#Find null values
data.isnull().sum()
```

Replacing missing values

```python
data.fillna(method ='pad')  # replacing with prev value
```

```python
# column
mean_value=data['temp'].mean()
  
# mean of values in the same column
data['temp'].fillna(value=mean_value, inplace=True)
print('Updated Dataframe:')
print(data)
```

```python
numerical_feature = data.describe(include=["int", "float"]).columns
print(data[numerical_feature])
```

```python
# Convert categorical variables (months and days) into numerical values
months_table = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
days_table = ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat']

data['month'] = [months_table.index(month) for month in data['month'] ]
data['day']   = [days_table.index(day)  for day in data['day']   ]
```

```python
data.info()
```

```python
data['impact'] = [0 if area == 0 else 1 for area in data['area']]
target = data.impact
```

```python
#standardization of data
scaler = StandardScaler()

t_df = data.drop(['area', 'impact'], axis =1)

#fitting forest fire dataset to scaler by removing the attribute output
scaler.fit(t_df)

scaled_features = scaler.transform(t_df)
# scaled_features = scaler.transform(data.drop('area',axis=1))
inputs = pd.DataFrame(scaled_features,columns=data.columns[:-2])
inputs.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.3, random_state=0)
```
